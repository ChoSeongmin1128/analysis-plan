# 1. 개요
## 분석의 목적
현재 공교육에 대한 문제를 찾아보던 중 사교육비가 매년 증가한다는 것을 확인하였고, 이로 인해 공교육의 영향력이 줄어들고 있을 수 있다는 의문을 해소하기 위함입니다.

## 문제 설명
교육 공공데이터를 활용하여 데이터의 목적에 부합한 해석 및 해결 방안 제시

# 2. 데이터 수집
## 데이터 출처
- 학원 및 교습소 데이터: 행정구역 별 교육청에서 제공
- 학업 성취도, 대학수학능력시험 데이터: EDSS
- 사교육비: 국가통계포털 KOSIS

## 수집 방법
- EDSS에 직접 요청을 통한 데이터 수집
- 공식 웹사이트(교육청, 국가통계포털, 데이터 포털)를 통해 접근 가능한 데이터 파일 다운로드

## 수집된 데이터의 구조 및 특성 설명
- 학원 및 교습소 데이터: 지역, 학원명, 학원종류, 분야, 교습 과목, 정원, 교습비 등
- 학업 성취도 데이터: 지역별, 학년별, 과목별 성적 백분위 및 평균 점수
- 대학수학능력시험 데이터: 지역별, 학년별, 과목별 성적 백분위 및 평균 점수
- 사교육비 데이터: 행정구역별, 학년별, 월평균 사교육비 및 사교육 참여율

# 3. 데이터 전처리
## 결측치 처리
- 데이터에 따라 평균값, 중앙값, 최빈값으로 대체 또는 결측치 제거

## 이상치 처리
- Z-score 또는 IQR 방법을 사용하여 이상치 식별 및 제거

## 데이터 정규화/표준화
- 학원 교습비 등 금액을 나타내는 데이터의 경우 다른 데이터에 비해 값이 크기 때문에 정규화 필요

## 피처 엔지니어링
- 미정

# 4. 데이터 탐색 및 시각화
## 기술통계 분석
- 각 변수의 기술 통계량(평균, 표준편차 등) 계산

## 데이터 분포 확인
- 각 변수의 분포를 히스토그램 또는 박스 플롯으로 시각화

## 상관관계 분석
- 학원 밀집도와 학업 성취도 사이의 상관 계수 계산
